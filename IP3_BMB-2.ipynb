```python
X = np.linspace(0, 100, 100)

X_squared = X**2

fig=plt.figure(figsize=(16,8))
ax1=fig.add_subplot(231)
ax1.set_title('X')
ax1.plot(X)

ax2=fig.add_subplot(232)
ax2.set_title('X Squared')
ax2.plot(X_squared)

plt.show()


```

```python
exp_X = np.exp(X)

plt.figure(figsize=(8, 4))
plt.scatter(X, exp_X, label='X vs exp(X)', color='g')
plt.xlabel('X')
plt.ylabel('exp(X)')
plt.title('Scatter Plot of X vs exp(X)')

plt.show()
```

Price didnt work bc its an object value but odometer is an int value so it worked.

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,]', '', regex=True).astype('int64')

car_sales.head()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.head()
```

```python
from datetime import datetime, timedelta
```

```python
periods = len(car_sales)

start_date = datetime.today()
date_range = pd.date_range(start=start_date, periods=periods)

car_sales['Sale Date'] = date_range

car_sales.head()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(car_sales['Total Sales'], car_sales['Price'])

plt.xlabel('Total Sales')
plt.ylabel('Price')
plt.title('Price vs. Total Sales')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Colour', y='Total Sales', data=car_sales)
plt.xlabel('Colour')
plt.ylabel('Total Sales')
plt.title('Box Plot of Colour vs. Total Sales')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lmplot(y='Odometer (KM)', x='Total Sales', data=car_sales, hue='Make',ci=False)
plt.xlabel('Odometer (KM)')
plt.ylabel('Total Sales')
plt.title('Linear Regression: Total Sales vs. Odometer (KM)')

```

```python
numeric_columns = car_sales.select_dtypes(include=['number'])
sns.pairplot(numeric_columns)
```

```python
selected_columns = car_sales[['Total Sales', 'Price', 'Odometer (KM)']]

# Calculate the Pearson correlation matrix
correlation_matrix = selected_columns.corr()

# Extract the Pearson correlation coefficient
pearson_correlation = correlation_matrix.loc['Total Sales', 'Price']

plt.show()
```

```python
heat_map=car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr(method='pearson')
print(heat_map)
cols=['Total Sales', 'Price', 'Odometer (KM)']
sns.heatmap(heat_map,xticklabels=cols,yticklabels=cols,annot=True)
```
